```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
```

```python
waterfall=imread('niagara_waterfall.jpg') 
```

```python
plt.figure(figsize=(14,10))  
plt.imshow(waterfall)
```

```python
waterfall
```

```python
waterfall=waterfall/255.0
```

```python
waterfall.shape
```

```python
Y = waterfall.reshape(-1,3) 
```

```python
Y.shape
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
km.fit(Y)
```

```python
image_seg_1=km.cluster_centers_
```

```python
image_seg_1=image_seg_1[km.labels_]
```

```python
image_seg_1=image_seg_1.reshape(waterfall.shape)
```

```python
plt.figure(figsize=(12,10))
plt.imshow(image_seg_1)
```

```python
fig = plt.figure(figsize=(10,10))
ax=fig.add_subplot(1,2,1,xticks=[],yticks=[],title="Original Image")
ax.imshow(waterfall)
ax=fig.add_subplot(1,2,2,xticks=[],yticks=[],title="Colour Compressed Image Using K-Means")
ax.imshow(image_seg_1)
plt.show()
```
